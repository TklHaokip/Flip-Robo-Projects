```python
#Libraries
import pprint
from gensim import corpora
from collections import defaultdict
from gensim.models import Word2Vec
```

```python
# Create a set of frequent words
stoplist = set('for a of the and to in are which use one should or more they have that also with when it as from you an first can be even if there is no go will than would few does so only not has by much but why'.split(' '))

# Lowercase each document,split it by white space and filter out stopwords
texts = [[word for word in doc.lower().split() if word not in stoplist]
        for doc in corpus]
```

```python
# Count word frequencies
frequency = defaultdict(int)

for text in texts:
    for token in text:
        frequency[token] += 1
        
# Only keep words that appear more than once
processed_corpus = [[token for token in text if frequency[token] >1] for text in texts]
pprint.pprint(processed_corpus)
```

```python
# Creating dictionaries,which helps during topic modelling
dictionary = corpora.Dictionary(processed_corpus)
print(dictionary)
```

```python
# Vector representation
pprint.pprint(dictionary.token2id)
```

```python
#Training the Word2Vec model
model = Word2Vec(processed_corpus,min_count=1)
words = model.wv.vocab
```

```python
# Test the word Vectors
vector = model.wv['vehicles']
print(vector)
```

```python
# Most similar words
similar = model.wv.most_similar('engine',topn=30)
similar
```

```python
# Check words similarity and compare above
model.wv.similarity(w1='fuel',w2='cost')
```

```python
model.wv.similarity(w1='charging',w2='power')
```

```python
# Preprocessing the data
text = re.sub(r'\[[0-9]*\]',' ',doc)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
```

```python
# Preparing the dataset
sentences = nltk.sent_tokenize(text)
sentences
```

```python
sent_word = [nltk.word_tokenize(sentence) for sentence in sentences]
sent_word
```

```python
# sent_word = [nltk.word_tokenize(sentence) for sentence in sentences]
for i in range(len(sent_word)):
    sent_word[i] = [word for word in sent_word[i] if word not in stopwords.words('english')]
```

```python
from sklearn.manifold import TSNE
vocab = ['social','creatures','fantasy','real','life','trigger','exceptionally','optimistic','entrepreneur',
         'stake','mistaken','generous','way','dreaming','bigger','learn','competitor','persistent','opportunity','impacted']
def tsne_plot(model):
    labels = []
    wordvecs = []
    
    for word in vocab:
        wordvecs.append(model[word])
        labels.append(word)
        
    tsne_model = TSNE(perplexity=3,n_components=3,init='pca',random_state=42)
    coordinates = tsne_model.fit_transform(wordvecs)
    
    x=[]
    y=[]
    for value in coordinates:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(15,8))
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                    xy=(x[i], y[i]),
                    xytext=(2,2),
                    textcoords='offset points', ha='left')
    
    plt.show()
    
tsne_plot(model)
```
